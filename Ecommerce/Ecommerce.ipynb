# Training and Testing Data

Now that we've explored the data a bit, let's go ahead and split the data into training and testing sets. Set a variable X equal to the numerical features of the customers and a variable y equal to the "Yearly Amount Spent" column.

```python
X = cust_data.iloc[:,:4]
```

```python
X
```

```python
y = cust_data['Yearly Amount Spent']
```

```python
y
```

Split the Data for model training 70% and testing 30% 

```python
from sklearn.model_selection import train_test_split 
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,y,test_size=0.3,random_state=101)
```

```python
print(f"train data shape: {X_train.shape} and {Y_train.shape}")
print(f"test data shape: {X_test.shape} and {Y_test.shape}")
```

# Train the model

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
model_1 = lm.fit(X_train,Y_train)
```

```python
print("slop:",model_1.coef_)
```

```python
print("intercept",model_1.intercept_)
```

# Prediction of trained data

```python
prediction_1 = model_1.predict(X_test)
```

```python
prediction_1 
```

# Create a scatterplot of the real test values versus the predicted values.

```python
plt.scatter(x = Y_test, y = prediction_1, color="Red");
plt.ylabel("predicted value")
plt.xlabel("Real value")
```

```python
print("Accuracy:{}%".format(model_1.score(X_test,Y_test)*100))
```

```python
sns.scatterplot(x=Y_test,y=prediction_1)
```

# Evaluation of model

Let's evaluate our model performance by calculating the residual sum of squares and the explained variance score (R^2).

Calculate the Mean Absolute Error, Mean Squared Error, and the Root Mean Squared Error.

```python
import sklearn.metrics as metrics
import numpy as np
```

```python
print(f"MAE:{metrics.mean_absolute_error(Y_test,prediction_1)}")
print(f"MSE:{metrics.mean_squared_error(Y_test,prediction_1)}")
print(f"RMSE:{np.sqrt(metrics.mean_squared_error(Y_test,prediction_1))}")
```

Plot a histogram of the residuals and make sure it looks normally distributed
The difference between what was expected and what was predicted is called the residual error. It is calculated as: residual error = expected - predicted.

```python
sns.histplot((Y_test-prediction_1));
```
