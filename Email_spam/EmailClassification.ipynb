```python
import pandas as pd
```

```python
main_data = pd.read_csv("E:\DATA-SCIENCE\Project\Email_spam\mail_data.csv")
```

**Binary Encoding**

```python
#Convert the spam as 0 and ham as 1 
main_data.loc[main_data['Category'] == "spam",'Category',] = 0 
main_data.loc[main_data['Category'] == "ham","Category",] = 1 
```

```python
main_data
```

```python
main_data.loc[:0] #return the 0 index value 
```

**Data Cleaning**

```python
main_data.isna().sum() #No Nun data found
```

```python
#Selection of Dependent  and independent variable 
X = main_data['Message']    #independent object
y = main_data['Category']   #dependent or target object
```

```python
#Split the data into training and testing part
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,y,test_size=0.2,random_state=3) 
```

```python
X_train.shape,X_test.shape  #4457 for train data vs 1115 for test data
```

# Model Development for text data

**here we use TfidfVectorizer library which is combination of Term Frequency and Inverse Document Frequency.**

tf shows us how many time the term(word) are repeated in string, and idf that will count the weight(importance) of that word from document.

```python
from sklearn.feature_extraction.text import TfidfVectorizer as tf
```

```python
vector_model = tf(min_df=1,stop_words="english",analyzer="word",lowercase=True)    
```

```python
#let fit-trasform the train data 
X_train_vector = vector_model.fit_transform(X_train)
```

```python
#let tranform the test data 
X_test_vector = vector_model.transform(X_test)
```

```python
print(X_train_vector)
```

In the context of text analysis and TF-IDF, this could represent the TF-IDF score for a particular term (word) in a document. The row index (0) could correspond to the document, the column index (5413) could correspond to the specific term in the vocabulary, and the value (0.6198254967574347) represents the TF-IDF score for that term in the document.

```python
#for example if we want to check whats inside the term of 5413 at 0 index 
feature_name = vector_model.get_feature_names_out()
term_5413 = feature_name[5413]
print(f"Here the term is:{term_5413}")
print("==================================")

#if we want to check whole term than 
for index,term in enumerate(feature_name):
    print(f"{index} : {term}")
```

```python
print(X_test_vector)
```

```python
#convert the datatype from pandas to numerical
Y_train = Y_train.astype('int')
Y_test = Y_test.astype('int')
```

# Model development for Classification 

```python
#We use logistic regression for this data
from sklearn.linear_model import LogisticRegression
```

```python
Model = LogisticRegression().fit(X_train_vector,Y_train)
```

**Prediction**

```python
#Prediction on traning data
Pred_train_data = Model.predict(X_train_vector)
print(Pred_train_data)
#Accuracy of traing data 
print(f"Accuracy on traning data:{Model.score(X_train_vector,Y_train)*100:.2f}")
```

```python
#Prediction of test data and accuracy score
from sklearn.metrics import accuracy_score
pred_test_data = Model.predict(X_test_vector)
Accuracy_score = accuracy_score(Y_test,pred_test_data)
print(f"Prediction on test data Accuracy score:{Accuracy_score*100:.2f}")
```

**USER INPUT PREDICTION**

```python
user_input = ['free free free click on link for free money']
input_data = vector_model.transform(user_input)        #transform the user data(test data)
pred_user_data = Model.predict(input_data)

if (pred_user_data[0] == 0):
    print("This Email is SPAM")
else:
    print("This Email is HAM")
```
