# model development

```python
cols = ['width','height','color_score']
x = main_data.loc[:,cols].values
```

```python
#checking p-values
from scipy.stats import pearsonr
for i in cols:
    corr_coef,p_value = pearsonr(a[i],a['fruit_label'])
    print(f"Correlation between {i} and fruit_label: {corr_coef:.4f}, p-value: {p_value:.4f}")
```

```python
StandardScaler?
```

#StandardScaler used to scale all features value in one scale so the model better understand the data. 
for example : like if we have diff measurement of height and widht in meter and centermeter then model givie more priority to centemeter beacause of its high value so here we use standardscaler to scale all the values in one form.

```python
sc = StandardScaler()
X = sc.fit_transform(x)
```

```python
#Split the dataset 
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test ,Y_train,Y_test = train_test_split(X,y,test_size=0.2,random_state=101)
```

```python
X_train.shape,X_test.shape
```

```python
Y_train.shape
```

```python
# model : let use decision treee for classification of data 
from sklearn.tree import DecisionTreeClassifier
```

```python
ds = DecisionTreeClassifier()
model = ds.fit(X_train,Y_train)
```

```python
print("Training data score:",model.score(X_train,Y_train)*100)
```

```python
cols
```

```python
import sklearn.tree as tree
import matplotlib.pyplot as plt
```

```python
#ploting the classification process, how our data classified in diff class
fig = plt.figure(figsize=(20,20))
tree.plot_tree(model,feature_names=cols,class_names=['Apple', 'Mandarin','Orange','Lemon']);
```

# User input prediction

```python
user_input = []
num_input = int(input("How many instances or FRUIT you want to predict:"))
for i in range(num_input):
    width = float(input(f"Enter the Width Value {i+1}:"))
    height = float(input(f"Enter the Height Value {i+1}:"))
    color_score = float(input(f"Enter the Color_score Value {i+1}:"))
    user_input.append([width,height,color_score])

#now transform scale the user input data as we have scaled data in model
scaled_user_input = sc.transform(user_input)
#predict the user input 
user_predict = model.predict(scaled_user_input)

#show the result with fruit name 
for i, (predictions,input_values) in enumerate(zip(user_predict,user_input)):
    fruit_name = main_data.loc[main_data['fruit_label'] == predictions,'fruit_name'].values[0]
    print(f"Instance {i+1}:Predicted fruit label:{predictions},Corresponding Fruit name:{fruit_name},Input Value:{input_values}")
```
